# Ex2 - Filtering and Sorting Data

This time we are going to pull data directly from the internet.

### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/Euro_2012_stats_TEAM.csv).

### Step 3. Assign it to a variable called euro12.

```python
euro12 = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/Euro_2012_stats_TEAM.csv')
display(euro12.head())
```

### Step 4. Select only the Goal column.

```python
display(euro12['Goals'])
```

### Step 5. How many team participated in the Euro2012?

```python
num_teams = euro12.shape[0]
print(f"Number of teams participated in Euro 2012: {num_teams}")
```

### Step 6. What is the number of columns in the dataset?

```python
num_columns = euro12.shape[1]
print(f"Number of columns in the dataset: {num_columns}")
```

### Step 7. View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline

```python
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
display(discipline.head())
```

### Step 8. Sort the teams by Red Cards, then to Yellow Cards

```python
discipline_sorted = discipline.sort_values(by=['Red Cards', 'Yellow Cards'])
display(discipline_sorted)
```

### Step 9. Calculate the mean Yellow Cards given per Team

```python
mean_yellow_cards = euro12['Yellow Cards'].mean()
print(f"The mean number of Yellow Cards per team is: {mean_yellow_cards}")
```

### Step 10. Filter teams that scored more than 6 goals

```python
teams_more_than_6_goals = euro12[euro12['Goals'] > 6]
display(teams_more_than_6_goals)
```

### Step 11. Select the teams that start with G

```python
teams_starting_with_g = euro12[euro12['Team'].str.startswith('G')]
display(teams_starting_with_g)
```

### Step 12. Select the first 7 columns

```python
first_7_columns = euro12.iloc[:, :7]
display(first_7_columns)
```

### Step 13. Select all columns except the last 3.

```python
all_except_last_3 = euro12.iloc[:, :-3]
display(all_except_last_3)
```

### Step 14. Present only the Shooting Accuracy from England, Italy and Russia

```python
shooting_accuracy = euro12.loc[euro12['Team'].isin(['England', 'Italy', 'Russia']), ['Team', 'Shooting Accuracy']]
display(shooting_accuracy)
```
