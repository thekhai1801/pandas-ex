### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv'
chipo = pd.read_csv(url, sep='\t')
```

### Step 4. How many products cost more than $10.00?

```python
chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)
products_over_10 = chipo[chipo['item_price'] > 10.00]
num_products_over_10 = products_over_10.shape[0]
print(f"Số lượng sản phẩm có giá lớn hơn $10.00 là: {num_products_over_10}")
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python
item_prices = chipo[['item_name', 'item_price']]
display(item_prices)
```

### Step 6. Sort by the name of the item

```python
chipo_sorted_by_name = chipo.sort_values(by='item_name')
display(chipo_sorted_by_name)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
most_expensive_item = chipo.loc[chipo['item_price'].idxmax()]
quantity_of_most_expensive = most_expensive_item['quantity']
print(f"Số lượng của mặt hàng đắt nhất được đặt là: {quantity_of_most_expensive}")
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
veggie_salad_bowl_orders = chipo[chipo['item_name'] == 'Veggie Salad Bowl']
num_veggie_salad_bowl_orders = veggie_salad_bowl_orders.shape[0]
print(f"Số lần món 'Veggie Salad Bowl' được đặt là: {num_veggie_salad_bowl_orders}")
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
canned_soda_orders = chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)]
num_canned_soda_orders = canned_soda_orders.shape[0]
print(f"Số lần ai đó đặt nhiều hơn một 'Canned Soda' là: {num_canned_soda_orders}")
```
