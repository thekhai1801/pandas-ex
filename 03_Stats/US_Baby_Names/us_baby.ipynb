# US - Baby Names

### Introduction:

We are going to use a subset of [US Baby Names](https://www.kaggle.com/kaggle/us-baby-names) from Kaggle.  
In the file it will be names from 2004 until 2014


### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/US_Baby_Names_right.csv).

### Step 3. Assign it to a variable called baby_names.

```python
baby_names = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/US_Baby_Names_right.csv')
```

### Step 4. See the first 10 entries

```python
display(baby_names.head(10))
```

### Step 5. Delete the column 'Unnamed: 0' and 'Id'

```python
baby_names = baby_names.drop(['Unnamed: 0', 'Id'], axis=1)
display(baby_names.head())
```

### Step 6. Is there more male or female names in the dataset?

```python
gender_counts = baby_names['Gender'].value_counts()
display(gender_counts)
```

### Step 7. Group the dataset by name and assign to names

```python
names = baby_names.groupby('Name').sum(numeric_only=True)
display(names.head())
```

### Step 8. How many different names exist in the dataset?

```python
display(len(names))
```

### Step 9. What is the name with most occurrences?

```python
most_common_name = names.sort_values(by='Count', ascending=False).head(1)
display(most_common_name)
```

### Step 10. How many different names have the least occurrences?

```python
least_common_count = names.sort_values(by='Count', ascending=True).head(1)['Count'].iloc[0]
least_common_names_count = names[names['Count'] == least_common_count].shape[0]
display(least_common_names_count)
```

### Step 11. What is the median name occurrence?

```python
median_occurrence = names['Count'].median()
display(median_occurrence)
```

### Step 12. What is the standard deviation of names?

```python
std_occurrence = names['Count'].std()
display(std_occurrence)
```

### Step 13. Get a summary with the mean, min, max, std and quartiles.

```python
display(names['Count'].describe())
```
