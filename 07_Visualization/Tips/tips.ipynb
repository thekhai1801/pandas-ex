### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv).

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv')
display(tips.head())
```

### Step 4. Delete the Unnamed 0 column

```python
tips = tips.drop('Unnamed: 0', axis=1)
display(tips.head())
```

### Step 5. Plot the total_bill column histogram

```python
sns.histplot(tips['total_bill'], kde=False)
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Histogram of Total Bill')
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs Tip')
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
sns.relplot(x='total_bill', y='tip', size='size', data=tips)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Relationship between Total Bill, Tip, and Size')
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
plt.title('Total Bill by Day of the Week')
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.scatterplot(x='tip', y='day', hue='sex', data=tips)
plt.xlabel('Tip')
plt.ylabel('Day of the Week')
plt.title('Tip Amount by Day and Sex')
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(x='day', y='total_bill', hue='time', data=tips)
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
plt.title('Total Bill by Day of the Week and Time')
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
dinner_tips = tips[tips['time'] == 'Dinner']['tip']
lunch_tips = tips[tips['time'] == 'Lunch']['tip']

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(dinner_tips, kde=False, ax=axes[0])
axes[0].set_title('Tip Distribution for Dinner')
axes[0].set_xlabel('Tip')
axes[0].set_ylabel('Frequency')

sns.histplot(lunch_tips, kde=False, ax=axes[1])
axes[1].set_title('Tip Distribution for Lunch')
axes[1].set_xlabel('Tip')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
male_data = tips[tips['sex'] == 'Male']
female_data = tips[tips['sex'] == 'Female']

fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=male_data, ax=axes[0])
axes[0].set_title('Male - Total Bill vs Tip by Smoker Status')
axes[0].set_xlabel('Total Bill')
axes[0].set_ylabel('Tip')

sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=female_data, ax=axes[1])
axes[1].set_title('Female - Total Bill vs Tip by Smoker Status')
axes[1].set_xlabel('Total Bill')
axes[1].set_ylabel('Tip')

plt.tight_layout()
plt.show()
```

### BONUS: Create your own question and answer it using a graph.

```python
tips['tip_percentage'] = (tips['tip'] / tips['total_bill']) * 100

avg_tip_percentage = tips.groupby(['day', 'time'])['tip_percentage'].mean().reset_index()

sns.barplot(x='day', y='tip_percentage', hue='time', data=avg_tip_percentage)
plt.xlabel('Day of the Week')
plt.ylabel('Average Tip Percentage (%)')
plt.title('Average Tip Percentage by Day and Time')
plt.show()
```
